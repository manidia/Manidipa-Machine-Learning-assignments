```python
'''Using Naive Bayes algorithms, predict if a person is diabetic or not, based on the features provided. Find the accuracy and F1-Scores of both algorithms.
Dataset: pima-indians-diabetes.data.csv
'''
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.metrics import accuracy_score, f1_score

# Load and prepare data
df = pd.read_csv('pima-indians-diabetes.data.csv', header=None)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

# Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
acc_gnb = accuracy_score(y_test, y_pred_gnb)
f1_gnb = f1_score(y_test, y_pred_gnb)

# Bernoulli Naive Bayes
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
y_pred_bnb = bnb.predict(X_test)
acc_bnb = accuracy_score(y_test, y_pred_bnb)
f1_bnb = f1_score(y_test, y_pred_bnb)

print(f"GaussianNB - Accuracy: {acc_gnb:.4f}, F1-Score: {f1_gnb:.4f}")
print(f"BernoulliNB - Accuracy: {acc_bnb:.4f}, F1-Score: {f1_bnb:.4f}")
```
